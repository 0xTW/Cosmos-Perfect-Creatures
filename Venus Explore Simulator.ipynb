```python
import random
import time
from IPython.display import clear_output

solutionsDict = {
    '1': 'Traditional Solution',
    '2': 'Distributed Power Generation Solution(Spring)',
    '3': 'Distributed Power Generation Solution(Compressed Air)'
}

class Machine:
    def __init__(self, solution):
        self.days = 1
        self.exploreProgress = 0.0
        self.machineType = solution
        self.eleStations = 0        
        
        # Traditional Solution
        if solution == '1':
            self.eleCapacity = 1100
            self.maxCapacity = 1100
            self.buildStationCost = 0
        # Distributed Power Generation Solution(Spring)
        elif solution == '2':
            self.eleCapacity = 220
            self.maxCapacity = 220
            self.buildStationCost = 200
        # Distributed Power Generation Solution(Compressed Air)
        elif solution == '3':
            self.eleCapacity = 440
            self.maxCapacity = 440
            self.buildStationCost = 150
        else:
            raise('Error Solutions.')
    
    def generateEle(self):
        if self.machineType == '1':
            return random.choice([0, 108, 216, 270])
        elif self.machineType == '2':
            acc = 0
            for i in range(self.eleStations):
                acc += random.choice([0, 5, 10, 11])
            return acc
        elif self.machineType == '3':
            acc = 0
            for i in range(self.eleStations):
                acc += random.choice([0, 30, 60, 65])
            return acc

class Action:
    def __init__(self, machine):
        self.machine = machine
    def Explore(self):
        if self.machine.eleCapacity > 0: 
            self.machine.exploreProgress += (self.machine.eleCapacity - 20) / 500
            self.machine.eleCapacity = 20
            self.DoNothing()
        else:
            print('Electronic is not enough to explore.')
    def Build(self):
        if self.machine.machineType == '1':
            print('Not thing happend.')
        elif self.machine.eleStations >= 3:
            print('The max stations is 3.')
        elif self.machine.eleCapacity >= self.machine.buildStationCost:
            self.machine.eleStations += 1
            self.machine.eleCapacity -= self.machine.buildStationCost
        else:
            print('Electronic is not enough to build.')
            
    def DoNothing(self):
        print('End Day.')
        self.machine.days += 1
        self.machine.eleCapacity += self.machine.generateEle()
        self.machine.eleCapacity -= 5
        self.machine.eleCapacity = min(self.machine.eleCapacity, self.machine.maxCapacity)
        if self.machine.machineType == '1' and self.machine.eleCapacity < 0:
            self.machine.eleCapacity = 0
        clear_output(wait=False)
        time.sleep(0.5)
```

```python
print('Please choose your machine type (solutions)')
print('1. Traditional Solution')
print('2. Distributed Power Generation Solution(Spring)')
print('3. Distributed Power Generation Solution(Compressed Air)')
solution = input('>')

if solution not in ('1', '2', '3'):
    raise('Error Solution.')
m = Machine(solution)

while True:
    print(f'Solution: {solutionsDict[m.machineType]}')
    print(f'Day {m.days}, ðŸ”‹Capacity: {m.eleCapacity}, ðŸ§­Explore: {round(m.exploreProgress, 2)}%, âš¡Power Stations: {m.eleStations}')
    
    if m.eleCapacity < 0 and m.machineType != '1':
        print('Machine is out of electronic, the emulator is done.')
        break
    if m.exploreProgress >= 100:
        print('Venus Explore is success, congrats.')
        break
    if m.days == 60:
        print('60 days exploration in Venus is over.')  
    
    print('Please Choose Actions:')
    if m.machineType != '1' or (m.machineType == '1' and m.eleCapacity > 5):
        print('1. Explore')
    if m.machineType != '1':
        print('2. Build Station')
    print('3. Day End')
    action = input('>')
    
    if action in (1, '1', 'e', 'explore'):
        Action(m).Explore()
    if action in (2, '2', 'b', 'build') and m.machineType != '1':
        Action(m).Build()
    if action in (3, '3', 'end'):
        Action(m).DoNothing()
    if action in (-1, '-1'):
        break
```
